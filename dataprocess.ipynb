```python
#normalize 可复用
def normalizeText(x):
    x = x.astype(float)
    Min = x.min()
    Max = x.max()
    x = (x - Min) / (Max - Min)
    return x
```

```python
listing_all['security_deposit_norm'] = normalizeText(listing_all['security_deposit1'])
print(listing_all['security_deposit_norm'] )
listing_all['cleaning_fee_norm'] = normalizeText(listing_all['cleaning_fee1'])
print(listing_all['cleaning_fee_norm'] )
```

```python
#重新将listing_all 中的host_response_time拼接回来
#此处注意！！本来正规做法是保留id之后进行merge拼接回来，但是因为操作麻烦，我们的数据
#也没有进行删除操作，所以index保持一致，将进行直接拼接,为防止后续进一步改进，同时拼接原始数据和改进后数据
listing_all = pd.read_csv('./listings_all.csv')
b['host_response_time'] = listing_all['host_response_time']
print('host_response_time:',listing_all['host_response_time'].unique())
listing_all['host_response_time'].fillna(0,inplace=True)
#3天
listing_all['host_response_time'].replace('a few days or more',72,inplace=True)
#1天
listing_all['host_response_time'].replace('within a day',24,inplace=True)
#半天
listing_all['host_response_time'].replace('within a few hours',12,inplace=True)
#1h
listing_all['host_response_time'].replace('within an hour',1,inplace=True)
#在最后一列
b['host_response_time_norm'] = normalizeText(listing_all['host_response_time'])
display(b.head(5))
```

```python
#以下两个格子根据晨阳给出的数据改写cancellation_policy
#在这个网站上找到了strict对应政策https://www.airbnb.com/home/cancellation_policies?locale=en#strict-with-grace-period
#发现strict等价于strict_14_with_grace_period
#近似处理方式，因为super_strict还有半返款限制，用is_super表示，1表示不能退全款，0表示可以
#！！！不要看这个地方输出的表格，是错误操作，下个block print的是正确的
b['is_super'] = b['cancellation_policy']
b['is_super'] = b['is_super'].apply(lambda x:0 if 'super' not in x else 1)
print('cancellation_policy:',b['cancellation_policy'].unique())
```

```python
b['cancellation_policy_norm'] = b['cancellation_policy']
#14
b['cancellation_policy_norm'] = b['cancellation_policy_norm'].replace('strict_14_with_grace_period',14)
b['cancellation_policy_norm'] = b['cancellation_policy_norm'].replace('strict',14)
#5
b['cancellation_policy_norm'] = b['cancellation_policy_norm'].replace('moderate',5)
#1
b['cancellation_policy_norm'] = b['cancellation_policy_norm'].replace('flexible',1)
#30
b['cancellation_policy_norm'] = b['cancellation_policy_norm'].replace('super_strict_30',30)
#60
b['cancellation_policy_norm'] = b['cancellation_policy_norm'].replace('super_strict_60',60)
b['cancellation_policy_norm'] = normalizeText(b['cancellation_policy_norm'])
display(b.head(5))
```

```python
#对host_is_superhost改写，把t&f改成1&0
b['host_is_superhost'] = b['host_is_superhost'].apply(lambda x:0 if 't' not in x else 1)
print(b['host_is_superhost'])
```

```python
#可复用string改binary方法
def binary_sparseMatrix(x):
    dstr = x
    dlist = x.tolist()
    category_column = x.unique().tolist()
    zero_array = pd.DataFrame(np.zeros((len(x),len(category_column))),columns=category_column)
    for m in range(x.shape[0]):
        zero_array.loc[m, dlist[m]] = 1
    return zero_array
```

```python
#对neighborhood_group改写
b=pd.concat([b,binary_sparseMatrix(b['neighbourhood_group'])],axis=1)
display(b.head(5))
```

```python
#更正room_type
b=pd.read_csv('./alldataAfterPreprocess.csv')
listing_all = pd.read_csv('./listings_all.csv')
b['room_type'] = listing_all['room_type']
b = pd.concat([b,binary_sparseMatrix(b['room_type'])],axis=1)
display(b.head(5))
```
